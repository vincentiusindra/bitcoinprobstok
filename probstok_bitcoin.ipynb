```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
pred_end_date = '30-4-2018'
register_matplotlib_converters()
prices = pd.read_csv('btc.csv', sep=';')
So = 13921.48064
dt = 1
T = 60
N = T / dt
t = np.arange(1, int(N) + 1)
returns = (prices.loc[1:, 'Harga'] - \
          prices.shift(1).loc[1:, 'Harga']) / \
          prices.shift(1).loc[1:, 'Harga']
mu = np.mean(returns)
sigma = np.std(returns)
scen_size = 2
b = {str(scen): np.random.normal(0, 1, int(N)) for scen in range(1, scen_size + 1)}
W = {str(scen): b[str(scen)].cumsum() for scen in range(1, scen_size + 1)}
drift = (mu - 0.5 * sigma**2) * t
diffusion = {str(scen): sigma * W[str(scen)] for scen in range(1, scen_size + 1)}
S = np.array([So * np.exp(drift + diffusion[str(scen)]) for scen in range(1, scen_size + 1)]) 
S = np.hstack((np.array([[So] for scen in range(scen_size)]), S))
print(S)
plt.figure(figsize = (20,10))
data = pd.read_csv("btc.csv", delimiter=';', index_col='Waktu')
data1 = pd.read_csv("prediksi.csv", delimiter=';', index_col='Waktu')
a = (data['Harga'][0:91])
a=a[1:len(a)]
b = (data1['Harga1'][0:60])
b=b[1:len(b)]
c = (data1['Harga2'][0:60])
c=c[1:len(c)]
plt.figure(figsize=(15,10))
plt.plot(a, label='Harga Bitcoin', linewidth=1.5, color="purple")
plt.plot(b, label='Prediksi 1', linewidth=1.5, color="black")
plt.plot(c, label='Prediksi 2', linewidth=1.5, color="orange")
plt.xticks(rotation='vertical')
plt.xlabel('Waktu')
plt.ylabel('Harga($)')
plt.title('Prediksi Harga Bitcoin')
plt.grid(True)
plt.legend()
plt.show()
```
